```python
import cv2
import os 
import numpy as np 
from matplotlib import pyplot as plt
import time 
import mediapipe as mp 
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():

        # Read feed
        ret, frame = cap.read()
        cv2.imshow('OpenCV Feed', frame)

        # Break gracefully
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
